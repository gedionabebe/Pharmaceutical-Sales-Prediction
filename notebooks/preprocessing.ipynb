```python
train_data['Date'] = datetime(train_data,'Date')
train_data.set_index('Date',inplace=False)
train_data.info()
```

```python
#train_data['StateHoliday'] = train_data['StateHoliday'].to_string()
train_data['StateHoliday'] = label_encode(train_data,'StateHoliday')
train_data.info()
```

```python
train_data['StateHoliday'].replace(0,'0',inplace=True)
```

```python
store_data['CompetitionDistance']=fill_ffill(store_data,'CompetitionDistance')
store_data['CompetitionOpenSinceYear']=fill_ffill(store_data,'CompetitionOpenSinceYear')
store_data['Promo2SinceWeek']=fill_bfill(store_data,'Promo2SinceWeek')
store_data['Promo2SinceYear']=fill_bfill(store_data,'Promo2SinceYear')
store_data['PromoInterval']=fill_bfill(store_data,'PromoInterval')
store_data.isna().sum()
```

```python
test_data.info()
```

```python
test_data['Date'] = datetime(test_data,'Date')
test_data['StateHoliday'] = label_encode(test_data,'StateHoliday')
test_data['Open']=fill_ffill(test_data,'Open')
test_data.info()
```

```python
train_data['Year'] =  train_data['Date'].apply(lambda x: x.year)
train_data['Month'] = train_data['Date'].apply(lambda x: x.month)
train_data['DayOfMonth'] = train_data['Date'].apply(lambda x: x.day)
```

```python
store_data.info()
```

```python
store_data['StoreType'] = label_encode(store_data,'StoreType')
store_data['Assortment'] = label_encode(store_data,'Assortment')
store_data['PromoInterval'] = label_encode(store_data,'PromoInterval')
store_data.info()
```

```python
train_data['is_month_end'] = train_data['Date'].dt.is_month_end
train_data['is_month_start'] = train_data['Date'].dt.is_month_start
```

```python
train_data['is_weekend'] = 0
train_data['is_weekend'] = np.where(train_data['DayOfWeek']>5,1,train_data['is_weekend'])
```

```python
train_data['is_month_end'] = label_encode(train_data,'is_month_end')
train_data['is_month_start']= label_encode(train_data,'is_month_start')
```

```python
test_data['Year'] =  test_data['Date'].apply(lambda x: x.year)
test_data['Month'] = test_data['Date'].apply(lambda x: x.month)
test_data['DayOfMonth'] = test_data['Date'].apply(lambda x: x.day)
test_data['is_month_end'] = test_data['Date'].dt.is_month_end
test_data['is_weekend'] = 0
test_data['is_weekend'] = np.where(test_data['DayOfWeek']>5,1,test_data['is_weekend'])
test_data['is_month_start'] = test_data['Date'].dt.is_month_start
test_data['is_month_end'] = label_encode(test_data,'is_month_end')
test_data['is_month_start']= label_encode(test_data,'is_month_start')
```

```python
merged_train_store_data = train_data.merge(store_data,on='Store')
merged_test_store_data = test_data.merge(store_data,on='Store')
```
